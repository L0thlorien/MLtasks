```python
y_train = train_df['pidAggr'].values
X_train = train_df.drop(['id','pidAggr','rep_cand','dem_cand'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=666)
```

```python
# stdsc = RobustScaler()
# X_train = stdsc.fit_transform(X_train)
# X_test = stdsc.transform(X_test)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
rbs_rfc_pipeline = Pipeline([('rbs', RobustScaler()),
                               ('rfc', RandomForestClassifier(n_estimators=100,random_state=666))])

rbs_rfc_pipeline.fit(X_train, y_train)
predictionRFC = rbs_rfc_pipeline.predict(X_test)
accuracy_score(y_test, predictionRFC)
```

```python
rbs_dtc_pipeline = Pipeline([('rbs', RobustScaler()),
                               ('dtc', DecisionTreeClassifier(class_weight='balanced', random_state=666))])
rbs_dtc_pipeline.fit(X_train, y_train)
predictionDTC = rbs_dtc_pipeline.predict(X_test)
accuracy_score(y_test, predictionDTC)

# decision_tree = DecisionTreeClassifier(class_weight='balanced')
# decision_tree.fit(X_train, y_train)
# Y_pred = decision_tree.predict(X_test)
# accuracy_score(y_test, Y_pred)
```

```python
rbs_knc_pipeline = Pipeline([('rbs', RobustScaler()),
                               ('knc', KNeighborsClassifier())])
rbs_knc_pipeline.fit(X_train, y_train)
predictionKNC = rbs_knc_pipeline.predict(X_test)
accuracy_score(y_test, predictionKNC)

# knn = KNeighborsClassifier()
# knn.fit(X_train, y_train)
# Y_pred = knn.predict(X_test)
# accuracy_score(y_test, Y_pred)
```

```python
rbs_svc_pipeline = Pipeline([('rbs', RobustScaler()),
                               ('svc', SVC(kernel='rbf',random_state=666))])
rbs_svc_pipeline.fit(X_train, y_train)
predictionSVC = rbs_svc_pipeline.predict(X_test)
accuracy_score(y_test, predictionSVC)

# svc = SVC(kernel='rbf',random_state=0)
# svc.fit(X_train, y_train)
# Y_pred = svc.predict(X_test)
# accuracy_score(y_test, Y_pred)
```

```python
rbs_lsvc_pipeline = Pipeline([('rbs', RobustScaler()),
                               ('svc', LinearSVC(random_state=666))])
rbs_lsvc_pipeline.fit(X_train, y_train)
predictionLSVC = rbs_lsvc_pipeline.predict(X_test)
accuracy_score(y_test, predictionLSVC)

# linear_svc = LinearSVC()
# linear_svc.fit(X_train, y_train)
# Y_pred = linear_svc.predict(X_test)
# accuracy_score(y_test, Y_pred)
```
